```python
import numpy as np
import math
import matplotlib.pyplot as plt
from random import randint, uniform,random
```

# Muestras para una distribucion normal 

```python
def distribucion_normal(b,n=12):
    sumatoria = 0
    array = []
    for _ in range(0,n):
        aleatorio = randint(-b,b)
        array.append(aleatorio)
        sumatoria = sumatoria + aleatorio
        
    return array,1/2*sumatoria 
```

# Muestras para una distribucion Triangular

```python

def distribucion_triangular(b,n=12):
    array = []
    for _ in range(0,n):
        aleatorio =math.sqrt(6)/2*randint(-b,b)+randint(-b,b)
        array.append(aleatorio)

    return array
```

```python
b = 20
n = 10000
datos,suma_ = distribucion_normal(b,n)
array_datos = distribucion_triangular(b,n)
```

```python
cuenta, cajas, ignorar = plt.hist(datos, b)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma')
plt.show()
```

```python
cuenta, cajas, ignorar = plt.hist(array_datos, b)
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma')
plt.show()
```
